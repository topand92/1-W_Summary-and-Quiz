```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
data = {
    'Epoch': [1, 2, 3],
    'Training Loss': [1.198500, 0.351900, 0.296500],
    'Validation Loss': [0.415357, 0.403031, 0.405311]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 시각화
plt.figure(figsize=(8, 6))
plt.plot(df['Epoch'], df['Training Loss'], marker='o', label='Training Loss', color='b')
plt.plot(df['Epoch'], df['Validation Loss'], marker='o', label='Validation Loss', color='r')
plt.title('Training and Validation Loss by Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(df['Epoch'])
plt.grid(True)
plt.legend(title='Loss Type')
plt.show()
```

```python
# 데이터 준비
data = {
    'ROUGE Metric': ['ROUGE-1', 'ROUGE-2', 'ROUGE-L'],
    'Checkpoint 1': [0.5860, 0.3699, 0.5860],
    'Checkpoint 2': [0.4493, 0.2173, 0.4328],
    'Checkpoint 3': [0.4389, 0.2165, 0.4229]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 시각화
df.set_index('ROUGE Metric').T.plot(kind='line', marker='o', figsize=(8, 6))
plt.title('ROUGE Scores by Checkpoint')
plt.xlabel('Checkpoint')
plt.ylabel('ROUGE Score')
plt.xticks([0, 1, 2], ['Checkpoint 1', 'Checkpoint 2', 'Checkpoint 3'])
plt.grid(True)
plt.legend(title='ROUGE Metric')
plt.show()
```
